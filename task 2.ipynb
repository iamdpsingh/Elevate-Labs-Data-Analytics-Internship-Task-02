```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "SampleSuperstore.csv"
df = pd.read_csv(file_path)
```

```python
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
# Sales and Profit by Region
region_summary = df.groupby("Region")[["Sales", "Profit"]].sum().sort_values("Sales", ascending=False)
region_summary.plot(kind='bar', color=["#1f77b4", "#ff7f0e"])
plt.title("Sales and Profit by Region")
plt.ylabel("Amount (USD)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Profit by Sub-Category and Category
category_subcat_summary = df.groupby(["Category", "Sub-Category"])[["Sales", "Profit"]].sum().reset_index()
palette = sns.color_palette("Set2")
sorted_subcats = category_subcat_summary.sort_values("Profit")["Sub-Category"]
sns.barplot(data=category_subcat_summary, x="Profit", y="Sub-Category", hue="Category", dodge=True, palette=palette, order=sorted_subcats)
plt.title("Profit by Sub-Category and Category")
plt.xlabel("Profit (USD)")
plt.ylabel("Sub-Category")
plt.legend(title="Category", loc="lower right")
plt.tight_layout()
plt.show()
```

```python
# Sales and Profit by State (Top 20)
state_summary = df.groupby("State")[["Sales", "Profit"]].sum().sort_values("Sales", ascending=False).head(20)
state_summary.plot(kind="barh", color=["#2ca02c", "#d62728"])
plt.title("Top 20 States by Sales and Profit")
plt.xlabel("Amount (USD)")
plt.ylabel("State")
plt.tight_layout()
plt.show()
```

```python
# Profit vs Discount (Scatter Plot)
sns.scatterplot(data=df, x="Discount", y="Profit", hue="Category", palette="Dark2", alpha=0.7)
plt.title("Profit vs Discount")
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.axhline(0, color='gray', linestyle='--')
plt.tight_layout()
plt.show()
```

```python
# Sales and Profit by Segment
segment_summary = df.groupby("Segment")[["Sales", "Profit"]].sum().reset_index()
segment_summary.set_index("Segment")[["Sales", "Profit"]].plot(kind='bar', color=["#17becf", "#bcbd22"])
plt.title("Sales and Profit by Segment")
plt.ylabel("Amount (USD)")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Sub-Categories Ranked by Profit (Top & Bottom)
subcat_summary = df.groupby("Sub-Category")[["Profit"]].sum().sort_values("Profit")
sns.barplot(x=subcat_summary.Profit, y=subcat_summary.index, palette="coolwarm")
plt.title("Sub-Categories Ranked by Profit")
plt.xlabel("Profit (USD)")
plt.ylabel("Sub-Category")
plt.axvline(0, color='black', linestyle='--')
plt.tight_layout()
plt.show()
```
